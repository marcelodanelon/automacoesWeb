```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

import time

options = webdriver.ChromeOptions()
options.add_argument("--incognito")

navegador = webdriver.Chrome(options=options)
navegador.get("https://maestrosistemas.movidesk.com/")

# Maximiza tela
navegador.maximize_window()

# Digita Login
navegador.find_element(By.XPATH,'/html/body/div[1]/section/div/div[1]/div/div[2]/div[2]/div[2]/form/div/div/div[1]/input[1]').send_keys("marcelo.danelon")
# Digita Senha
navegador.find_element(By.XPATH,'/html/body/div[1]/section/div/div[1]/div/div[2]/div[2]/div[2]/form/div/div/div[2]/input[1]').send_keys("********")
time.sleep(3.0)
# Clica para entrar no sistema
navegador.find_element(By.XPATH,'/html/body/div[1]/section/div/div[1]/div/div[2]/div[2]/div[2]/form/div/div/div[4]/button').send_keys(Keys.ENTER)
time.sleep(5.0)
# Clica na opção de Relatórios
navegador.find_element('xpath','/html/body/aside/section/div[2]/a').click()
time.sleep(5.0)
# Clica no relatório
navegador.find_element('xpath','/html/body/section[1]/div[3]/div/div[1]/div[1]/div/div[2]/div[2]/ul/li[13]/a').click()
time.sleep(10.0)
# Clica em 'Exportar'
navegador.find_element('xpath','/html/body/section[1]/div[3]/div/div[1]/div[2]/div[1]/div[1]/div/div/div/div[2]/div[2]/button/span[1]').click()
time.sleep(5.0)
# Clica na opção 'Excel'
navegador.find_element('xpath','/html/body/section[1]/div[3]/div/div[1]/div[2]/div[1]/div[1]/div/div/div/div[2]/div[2]/ul/li[4]/a').click()
time.sleep(5.0)
# Clica no 'Ok' para concluir a exportação
navegador.find_element('xpath','/html/body/footer/div/div/div[3]/div/button[2]').click()
time.sleep(10.0)

navegador.quit()
```
